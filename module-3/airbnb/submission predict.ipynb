## Training

#### Random Forest

```python
X_train=df_train.drop("price",axis=1)
y_train=df_train.price
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR
rf=RFR()
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

#### LinReg

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train, y_train)
linreg.score(X_train, y_train)
```

#### CatBoost

```python
from catboost import CatBoostRegressor as CTR
ctr=CTR()
ctr.fit(X_train, y_train)
ctr.score(X_train, y_train)
```

#### Stacking from sklearn

```python
from sklearn.ensemble import GradientBoostingRegressor as GBR
gbr=GBR(min_samples_leaf=2,min_samples_split=4,n_estimators=200)

from xgboost import XGBRegressor as XGBR
xgbr=XGBR()

from lightgbm import LGBMRegressor as LGBMR
lgbmr=LGBMR()
```

```python
from mlxtend.regressor import StackingRegressor
metamodelo=StackingRegressor(regressors=[xgbr, gbr, lgbmr, linreg], meta_regressor=ctr)
metamodelo.fit(X_train, y_train)
metamodelo.score(X_train, y_train)
```
